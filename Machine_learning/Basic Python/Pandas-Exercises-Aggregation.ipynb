### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv("Exercise Data Files/Pandas Exercise Aggregation Files/occupation.csv", sep = '|')
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby(by = ['occupation'])[['age']].mean().add_prefix('Mean_Of_').reset_index()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
user_min = users.groupby(by = ['occupation','age'])
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
max_age = users.groupby(by = ['occupation'])[['age']].max().reset_index()
```

```python
min_age = users.groupby(by = ['occupation'])[['age']].min().reset_index()
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(by = ['occupation','gender'])[['age']].mean().reset_index()
```

### Step 8.  For each occupation present the percentage of women and men

```python
x=users.groupby(['occupation','gender'])['gender'].count()
```

```python
y = users.groupby(['occupation'])['gender'].count()
```

```python
r = ((x/y)*100)
```
